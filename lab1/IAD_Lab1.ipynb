Нехай n - кiлькiсть альтернативних моделей, m - кiлькiсть показникiв якостi, за якими цi моделi оцiнюються. 

Задано матрицю W значень прiоритетiв або величин виконання альтернатив моделей за показниками, де $w_{i,j} \in [0, 1]$  - нормований прiоритет (величина виконання) i-ої моделi
за j-м показником так що: 

$\sum_{i=1}^{n} w_{i,j} = 1 $ для кожного j = 1, ..., m. 

Задано також вектор нормованих ваг показникiв якостi $w_{j}^{q} \in [0, 1]$,  $\sum_{j=1}^{m} w_{j}^{q} = 1 $ 

Зобразити графiчно для кожної пари (i, k) альтернативних моделей у виглядi дiаграми, див. приклад нижче:

• значення рiзниць прiоритетiв $w_{i,j} $ − $w_{k,j} $ обраних моделей за кожним j-м показником якостi

• значення рiзницi агрегованих прiоритетiв (глобальних ваг)
$w_{i}^{aggr}- w_{k}^{aggr} $ обраних моделей, $w_{i}^{aggr} = \sum_{j=1}^{m}  w_{i,j}  w_{j}^{q}$

Таким чином, можемо вiзуально оцiнити вiдмiнностi мiж виконанням
двох обраних моделей за кожним показником якостi, а також якими є
вiдмiнностi у результатах - агрегованих прiоритетах цих двох моделей.
Кiлькiсть рисункiв з дiаграмами дорiвнює кiлькостi пар альтернативних
моделей.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from itertools import combinations
```

```python
n = 3
m = 3

w = np.array([[0.2,0.3, 0.4],
              [0.4, 0.3, 0.2],
              [0.3,0.2, 0.1],
              [0.1, 0.2, 0.3]])
w_q = np.array([0.2, 0.3, 0.5])
```

```python
w.shape
```

```python
list(combinations([i for i in range(w.shape[0])], 2))
```

```python
def difference(a, b):
  return a-b
```

```python
def aggr_diff(a, b,q):
  a_aggr = a*q
  b_aggr = b*q
  return a_aggr-b_aggr
```

```python
def plot_model_comparison(n, m, w, w_q):
  for i, j in list(combinations([i for i in range(w.shape[0])], 2)):
    df = pd.DataFrame({'w_'+str(i):w[i], 'w_'+str(j):w[j], 'diff':difference(w[i], w[j]),'aggr_diff':aggr_diff(w[i], w[j], w_q)})
    df.plot.barh()

    plt.title('Рiзницевий аналiз з 3 критерiями рiшень та 2 альтернативами')
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.show()
```

```python
plot_model_comparison(n, m, w, w_q)
```
